```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/drive/MyDrive/ECE 143 project/timesData.csv')
df= df.replace('-', 0)
df=df.fillna(value=0, axis=None, inplace=False, limit=None, downcast=None)
df['world_rank']=df['world_rank'].str.split("-").str[0]
df['world_rank'] = df['world_rank'].map(lambda x: x.lstrip('='))
df['world_rank']= df['world_rank'].astype('int')
df['income']=df['income'].astype(float)
df['international_students']=df['international_students'].str.split('%').str[0]

df_2011=df[df.year==2011]
df_2012=df[df.year==2012]
df_2013=df[df.year==2013]
df_2014=df[df.year==2014]
df_2015=df[df.year==2015]
df_2016=df[df.year==2016]

for i in range(200,200+len(df_2012)):
    df.loc[i,"world_rank"] = int(i-199)
for i in range(602,602+len(df_2013)):
    df.loc[i,"world_rank"] = int(i-601)
for i in range(1002,1002+len(df_2014)):
    df.loc[i,"world_rank"] = int(i-1001)
for i in range(1402,1402+len(df_2015)):
    df.loc[i,"world_rank"] = int(i-1401)
for i in range(1803,1803+len(df_2016)):
    df.loc[i,"world_rank"] = int(i-1802)

df_2011=df[df.year==2011]
df_2012=df[df.year==2012]
df_2013=df[df.year==2013]
df_2014=df[df.year==2014]
df_2015=df[df.year==2015]
df_2016=df[df.year==2016]

n_usa_top_100=[]
for i in range(2011,2017):
    n_usa_top_100.append(df[df.year==i]['country'][:100][df.country=='United States of America'].count())

n_germany_top_100=[]
for i in range(2011,2017):
    n_germany_top_100.append(df[df.year==i]['country'][:100][df.country=='Germany'].count())

ucsd_rankings=list(df['world_rank'][df['university_name']=='University of California, San Diego'])
ucsd_rankings.append(34) #2021 ranking
ucsd_details=df[df['university_name']=='University of California, San Diego']

```

```python
income_rank_2011=df_2011.sort_values(by='income',ascending=False)[['country','university_name','income']]
income_rank_2012=df_2012.sort_values(by='income',ascending=False)[['country','university_name','income']]
income_rank_2013=df_2013.sort_values(by='income',ascending=False)[['country','university_name','income']]
income_rank_2014=df_2014.sort_values(by='income',ascending=False)[['country','university_name','income']]
income_rank_2015=df_2015.sort_values(by='income',ascending=False)[['country','university_name','income']]
income_rank_2016=df_2016.sort_values(by='income',ascending=False)[['country','university_name','income']]

usa_univ_income=[]
usa_univ_income.append(income_rank_2011[:100][income_rank_2011['country']=='United States of America']['country'].count()/n_usa_top_100[0])
usa_univ_income.append(income_rank_2012[:100][income_rank_2012['country']=='United States of America']['country'].count()/n_usa_top_100[1])
usa_univ_income.append(income_rank_2013[:100][income_rank_2013['country']=='United States of America']['country'].count()/n_usa_top_100[2])
usa_univ_income.append(income_rank_2014[:100][income_rank_2014['country']=='United States of America']['country'].count()/n_usa_top_100[3])
usa_univ_income.append(income_rank_2015[:100][income_rank_2015['country']=='United States of America']['country'].count()/n_usa_top_100[4])
usa_univ_income.append(income_rank_2016[:100][income_rank_2016['country']=='United States of America']['country'].count()/n_usa_top_100[5])
plt.figure()
plt.plot(years,usa_univ_income)
plt.xlabel('Year')
plt.ylabel('Ratio of collaboration')
plt.title('Measure of industry collaboration by US universities')
plt.show()
```
